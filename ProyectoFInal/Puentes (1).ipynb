# Objetivos <a class="anchor" id="inicio"></a>

En este notebook se irá describiendo el paso a paso de la solución para los puentes de Konigsberg, el objetivo principal es encontrar al menos una fórmula en la que sea realizable el ejercicio. Dado el mapa de Königsberg, el río Prägel, que divide el plano en cuatro regiones distintas conectadas por siete puentes. Se quiere demostrar que camino(s) son posibles, comenzando a caminar desde cualquiera de estas regiones cruzando todos los puentes una sola vez y volviendo al mismo punto de partida.

# ¿Qué sucede en los puentes de  Königsberg? <a class="anchor" id="sat"></a>

 El problema se formuló en el siglo XVIII y consistía en encontrar un recorrido para cruzar a pie toda la ciudad pasando solo una vez por cada uno de los puentes y regresando al mismo punto de inicio.

Después de varios ensayos e investigación llegamos a resolver que es algo imposible, es necesario cruzar algún puente más de una vez. A base de repetir y repetir acabamos dándonos cuenta de que es un problema irresoluble. Ya que no existía ninguna manera en que se pudiesen atravesar los 7 puentes originales.

```python

#CAMBIAR LAS LETRAS POR NUMEROS, Y PONER PARA QUE SE MUESTRE CADA REGLA Y LUEGO SI LA TABLA

from Logica import *
from time import time
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from random import uniform, choice

def clasifica_para_tableaux(self):
    if type(self) == Letra:
        return None, 'literal'          # Literal positivo
    elif type(self) == Negacion:
        if type(self.subf) == Letra:
            return None, 'literal'      # Literal negativo
        elif type(self.subf) == Negacion:
            return 1, 'alfa'            # Fórmula 1 alfa
        if  self.subf.conectivo == 'O':
            return 3, 'alfa'
        elif self.subf.conectivo == '>':
            return 4, 'alfa'
        else:
            return 1, 'beta'

    elif type(self) == Binario:
        if self.conectivo == 'Y':
            return 2, 'alfa'
        elif self.conectivo == 'O':
            return 2, 'beta'
        elif self.conectivo == '>':
            return 3, 'beta'

setattr(Formula,"clasifica_para_tableaux",clasifica_para_tableaux)


# 1, alfa
print('Regla 1')
print(regla1.clasifica_para_tableaux())

# 2, alfa
print('Regla 2')
print(regla2.clasifica_para_tableaux())

 # 3, alfa
print('Regla 3')
print(regla3.clasifica_para_tableaux())

# 4, alfa
print('Regla 4')
print(regla4.clasifica_para_tableaux())

Ureglas = Binario('Y', regla1, regla2)
Ureglas_1 = Binario('Y', Ureglas, regla3)
UreglasF = Binario('Y', Ureglas_1, regla4)
print(UreglasF)

#regla1 = '((((1O2)Y(3O4))Y((5O6)Y(7O8)))Y(9O0))'
#regla2 = (((((1>(-1Y-2))Y(2>(-1Y-2)))Y((4>(-4Y-3))Y(3>(-4Y-3))))Y(((6>(-6Y-5))Y(5>(-6Y-5)))Y((8>(-8Y-7))Y(7>(-8Y-7)))))Y((9>(-9Y-0))Y(0>(-9Y-0))))'
#regla3 = '(((1O8)O9)Y((6O4)O8))'
#regla4 = '(((((1>4)Y(8>(3O5)))Y((9>6)Y(2>(8O9))))Y(((0>(8O1))Y(7>(1O9)))Y((4>(7O5))Y(3>2))))Y((6>(7O3))Y(5>0)))'
'''forms = []
forms = [inorder_to_tree(f) for f in forms]
n = nodos_tableaux(forms)
hijos = n.expandir()
print(hijos[0])'''

def primero_anchura(nodo):
    estado = nodo
    res = estado.es_hoja()
    if res == 'cerrada':
        return None
    elif res == 'abierta':
        return estado.interp()
    frontera = [estado]
    while len(frontera) > 0:
        estado = frontera.pop(0) 
        hijos = estado.expandir()
        for a in hijos:
            if a != None:
                res = a.es_hoja()
                if res == 'abierta':
                    return a.interp()
                elif res == None:
                    frontera.append(a)
    return None

forms = ['(((((((1O2)Y(3O4))Y((5O6)Y(7O8)))Y(9O0))Y(((((1>(-1Y-2))Y(2>(-1Y-2)))Y((4>(-4Y-3))Y(3>(-4Y-3))))Y(((6>(-6Y-5))Y(5>(-6Y-5)))Y((8>(-8Y-7))Y(7>(-8Y-7)))))Y((9>(-9Y-0))Y(0>(-9Y-0)))))Y(((1O8)O9)Y((6O4)O8)))Y(((((1>4)Y(8>(3O5)))Y((9>6)Y(2>(8O9))))Y(((0>(8O1))Y(7>(1O9)))Y((4>(7O5))Y(3>2))))Y((6>(7O3))Y(5>0))))']
forms = [inorder_to_tree(f) for f in forms]
n = nodos_tableaux(forms)
I = primero_anchura(n)
print()
print('Por anchura')
print("Resultado:", I)

def primero_profundidad(nodo):
    estado = nodo
    res = estado.es_hoja()
    if res == 'cerrada':
        return None
    elif res == 'abierta':
        return estado.interp()
    frontera = [estado]
    while len(frontera) > 0:
        estado = frontera.pop()
        hijos = estado.expandir()
        for a in hijos:
            if a != None:
                res = a.es_hoja()
                if res == 'abierta':
                    return a.interp()
                elif res == None:
                    frontera.append(a)
    return None
forms = ['(((((((1O2)Y(3O4))Y((5O6)Y(7O8)))Y(9O0))Y(((((1>(-1Y-2))Y(2>(-1Y-2)))Y((4>(-4Y-3))Y(3>(-4Y-3))))Y(((6>(-6Y-5))Y(5>(-6Y-5)))Y((8>(-8Y-7))Y(7>(-8Y-7)))))Y((9>(-9Y-0))Y(0>(-9Y-0)))))Y(((1O8)O9)Y((6O4)O8)))Y(((((1>4)Y(8>(3O5)))Y((9>6)Y(2>(8O9))))Y(((0>(8O1))Y(7>(1O9)))Y((4>(7O5))Y(3>2))))Y((6>(7O3))Y(5>0))))']
forms = [inorder_to_tree(f) for f in forms]
i = nodos_tableaux(forms)
L = primero_profundidad(i)
print()
print('Por profundidad')
print("Resultado:",L)

def backtracking(nodo):
    estado = nodo
    res = estado.es_hoja()
    if res == 'cerrada':
        return None
    elif res == 'abierta':
        return estado.interp()
    for hijo in estado.expandir():
        if hijo != None:
            resul = backtracking(hijo)
        if resul != None:
            return resul
    return None
forms = ['(((((((1O2)Y(3O4))Y((5O6)Y(7O8)))Y(9O0))Y(((((1>(-1Y-2))Y(2>(-1Y-2)))Y((4>(-4Y-3))Y(3>(-4Y-3))))Y(((6>(-6Y-5))Y(5>(-6Y-5)))Y((8>(-8Y-7))Y(7>(-8Y-7)))))Y((9>(-9Y-0))Y(0>(-9Y-0)))))Y(((1O8)O9)Y((6O4)O8)))Y(((((1>4)Y(8>(3O5)))Y((9>6)Y(2>(8O9))))Y(((0>(8O1))Y(7>(1O9)))Y((4>(7O5))Y(3>2))))Y((6>(7O3))Y(5>0))))']
forms = [inorder_to_tree(f) for f in forms]
i = nodos_tableaux(forms)
L = backtracking(i)
print()
print('Por backtracking')
print("Resultado:",L)

def obtiene_tiempos(fun, args, num_it=100):
    tiempos_fun = []
    for i in range(num_it):
        arranca = time()
        x = fun(*args)
        para = time()
        tiempos_fun.append(para - arranca)
    return tiempos_fun

def compara_entradas_funs(funs, nombres_funs, lista_args, N=100):
    entradas = []
    funcion = []
    tiempos = []
    lista_dfs = []
    for i, args in enumerate(lista_args):
        for j, fun in enumerate(funs):
            t = obtiene_tiempos(fun, [args], N)
            tiempos += t
            n = len(t)
            entradas += [i+1]*n
            funcion += [nombres_funs[j]]*n
        df = pd.DataFrame({'Long_entrada':entradas, 
                           'Funcion':funcion,
                           'Tiempo_prom':tiempos})
        lista_dfs.append(df)
    df = pd.concat(lista_dfs)
    sns.lineplot(x='Long_entrada',y='Tiempo_prom',hue='Funcion',data=df)
    plt.show()
    
cantidad = 20
letras = [chr(i) for i in range(256, 256+cantidad)]
lista = [letras[0]]
formula = letras[0]
for p in letras[1:]:
    neg1 = '-' if uniform(0,1) > .5 else ''
    neg2 = '-' if uniform(0,1) > .5 else ''
    conectivo = choice(['Y','O','>'])
    formula = neg2 + "(" + formula + conectivo + neg1 + p + ")"
    lista.append(formula)

nombres = ['primero anchura', 'primero profundidad', 'backtracking']
anchura = lambda formula: primero_anchura(nodos_tableaux([inorder_to_tree(formula)]))
profundidad = lambda formula: primero_profundidad(nodos_tableaux([inorder_to_tree(formula)]))
backtrack = lambda formula: backtracking(nodos_tableaux([inorder_to_tree(formula)]))
funs = [anchura, profundidad, backtrack]
compara_entradas_funs(funs, nombres, lista)
```
